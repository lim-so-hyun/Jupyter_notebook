  처음에 1월 2월 그래프를 봤을 때 형태가 너무 똑같이 나와서 코드가 잘못된줄 알았는데 자세히 보니 최댓값도 달라져있더라구요.

*8월, 9월, 10월

  그리고 소현님이 말씀하신 하차인원그래프가 승차인원 그래프에 비해 살짝 밀리는 게 너무 신기하네요! 그와중에 중간 부분( 12시부터 14   시부분. U자형에서 평평한부분)은 왜 승차인원 그래프랑 하차인원 그래프가 물리는 건지도 궁금하네요.. 처음엔 인원수때문에 이렇게 나타   나나 생각했어요. 피크를 찍는 부분은 유동인구가 많은만큼 승하차인원 차이가 분명하게 나타나고 중간부분은 적으니까 차이가 불분명하다   고 생각했어요. 근데 최소부분(양끝)을 보면 또 그래프가 물리지 않아서... 흠 어떻게 생각하시는지요.. 아무래도 점심시간쯤이니까 멀리   이동하기 힘드니 거리상, 시간상 차이가 별로 안나서 그래프가 물릴 수 있겠다는 생각도 들어요!
  

*보완
  :코로나를 중심으로 승하차인원을 비교해보았는데, 시각적으로 큰 차이를 볼 수 없어서 조금 더 세부적으로 분석하면 결과가 달라질지        궁금하네요. 
  :코로나라는 큰 영향요인 안에도 되게 다양한 요인이 담겨있기 때문에 짧은 시간안에 분석하기에는 조금 무리가 있는 것 같기도 해요. 기    회가 된다면 오래동안 연구하기 좋은 주제라는 생각도 드네요. 
  (코로나 격상 시기, 같은 격상 시기라도 대응책이 어떻게 달라졌는지, 집단 유행 시기, 그에 따른 사람들의 위기 의식 수준,    사람들이   코로나를 일상으로 받아들이는 정도 등등) 생각보다 매우 다양한 분야(심리, 법, 사회) 의 접목이 필요해보여요
  


```python
import pandas as pd
```

```python
sub20=pd.read_csv('subway.csv', encoding='cp949')
```

```python
sub20.drop('작업일자', axis='columns', inplace=True)
```

```python
sub20
```

```python
sub2001=sub20.loc[sub20['사용월']<202002] # 국내첫확진자발생일 1.21이지만 이 csv파일은 월로 구분되어있어 1월과 2월 비교
```

```python
sub2002=sub20.loc[(sub20['사용월']<202003) & (sub20['사용월']>202001)]
```

```python
sub2003=sub20.loc[(sub20['사용월']<202004) & (sub20['사용월']>202002)]
```

```python
sub2008=sub20.loc[(sub20['사용월']<202009) & (sub20['사용월']>202007)]
```

```python
sub2009=sub20.loc[(sub20['사용월']<202010) & (sub20['사용월']>202008)]
```

```python
sub2010=sub20.loc[(sub20['사용월']<202011) & (sub20['사용월']>202009)]
```

```python
stat2001 = sub2001.describe()
```

```python
# 2020.01. 시간대별 전역의 승차 및 하차 인원 평균
on_mean = stat2001.iloc[1, 1::2]
off_mean = stat2001.iloc[1, 2::2]
pd.options.display.float_format = '{:.2f}'.format  # 소숫점 둘째 자리까지만 보이게 하는 코드 추가
time = ['04시', '05시', '06시', '07시', '08시', '09시', '10시', '11시', '12시', '13시', '14시', '15시', '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시', '00시', '01시', '02시', '03시']
on_mean.index = time # 승차인원 평균
off_mean.index = time # 하차인원 평균

mean = pd.concat([on_mean, off_mean], axis=1) #승하차 같이 보기 위해 합침
mean.columns = ['승차인원 평균', '하차인원 평균']
mean = mean.transpose()
mean
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.title('2020년 1월의 시간대별 승하차인원')
plt.plot(time, on_mean, label='승차인원', color='skyblue')
plt.plot(time, off_mean, label='하차인원', color= 'darkblue')
plt.legend()
plt.show()
plt.savefig('2020년 1월의 시간대별 승하차인원.png')
```
