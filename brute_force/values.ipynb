# 학력별 일 vs 여가 선호도
* 전 학력에서 그래프 개형이 유사하게 나타났다. 왼쪽이 '일'을 상대적으로 더 중요시한다는 응답이고, 오른쪽이 '여가'를 중요시한다는 응답인데 그래프 개형을 살펴보면 모든 학력에서 그래프가 오른쪽으로 치우친 모습을 볼 수 있다.
* 모두 '일'보다는 '여가'를 중요시함을 확인할 수 있다.
* 세부적으로 학력별 분석을 하기에 앞서, 해당 통계조사의 조사대상을 확인해본 결과, '일반가구에 거주하는 만18-34세 이하 청년 3,500명 내외'로 정의되고 있었음. 즉, 애초에 조사 범위가 '청년'이기 때문에 사실상 이 데이터를 가지고 전연령대의 선호를 확인하기에는 어렵다.

* 학력에 따른 일 vs 여가 선호도를 조금 더 자세히 살펴보면, 그나마 '일'에 대한 선호도가 높게 나타나는 연령대는 '고졸 이하'였다. 그런데 면밀히 살펴보면 사실 이 네가지 범주를 함께 비교하는 것이 조금 무리가 있다고 생각된 것이 네 가지 범주 사이에 겹치는 부분이 존재하기 떄문이다. '졸업'을 기준으로 하고 있는 '고졸 이하', '대졸이상'에는 사실 '고졸이하'라고 했을 때에 '중고등학생'외에도 다양한 이유로 대학 진학을 하지 않은 청년들까지 포함을 하고 있다. '대졸이상'이라고 하면 '대학원생'도 포함되겠지만, 현재 학교에 재학중이지 않은 수많은 청년들을 포함하고 있다. 즉, 중고등학생과 대학생, 대학원생을 한 카테고리로, 그리고 '고졸이하'와 '대졸이상'을 한 카테고리로 보면 조금 더 다양한 해석이 가능할 것 같다. 

1) 중고등학생, 대학생 및 대학원생을 비교했을 때에는 확실히 파란색(중고등학생)이 '여가'에 대한 선호도가 매우 높고 '대학생, 대학원생'이 상대적으로는 '일'에 대한 선호도가 높은것으로 나타난다. 이는 아마도 '중고등학생'에게 있어 '일'은 생계를 위한 아르바이트, 직장생활보다는 '학업'일 것이므로 공부에 지쳐 '여가'를 희망하는 비율이 높은 것이 아닐까라고 생각한다. 반면, 대학생과 대학원생은 이제 학업이 아닌 '생계'를 위한 아르바이트가 일이 되기 때문에 일을 하면 본인이 쓸 수 있는 돈이 생긴다거나 사회적 지위가 올라간다거나 하는 부차적인 이득들이 생겨서 상대적으로 일에 대한 선호도가 '중고등학생'보다는 높아진 것이 아닐까 생각이 든다.  

2) 고졸 이하와 대졸이상을 비교하면 '고졸이하'에는 중고등학교 재학생 및 고등학교 졸업 이후 대학 진학을 하지 않은 청년들이 속하고, '대졸이상'은 대학 졸업 이후 대학원에 진학하는 학생들 혹은 취직하는 청년들이 포함되어있다. 이 둘 중에서는 '고졸 이하'가 상대적으로 '일'에 대한 선호도가 '대졸이상'보다 높게 나타나는데, 이는 '중고등학생'과 '대학생,대학원생'을 비교했을 때와 반대되는 결과이다. 
- 이렇게 반대되는 결과를 낳은 원인을 생각해보자면 아무래도 '대학에 진학하지 않은 청년들'이 기여하지 않았을까 하는 생각이 들었다. 중고등학생은 동일하게 포함이 되어있었는데 갑자기 '고졸 이하'의 일에 대한 선호도가 높아진 것은 다른 계층의 유입이 영향을 주었을 것이라고 판단하였기 떄문이다. 
- 고등학교 졸업 이후 다양한 이유로 '대학생'이 되지 않은 청년들의 경우 아무래도 바로 생계 등을 위한 취직 등에 뛰어드는 경우가 많을 것이다. 다양한 이유가 있을 것이라고 생각해서 함부로 단언하기는 어렵지만 아무래도 바로 취업 전선에 뛰어든 청년들이라면 '여가'에 대한 생각보다는 돈을 벌어 무언가를 성취하고자 하는 열망 혹은 의지가 있는 것으로 생각되어서 '일'에 대한 높은 선호도가 나타난 것이 아닐까 한다.
- 반면, '대졸이상'의 경우 '중고등학생'보다는 일에 대한 높은 선호도를 보이긴하지만, '고졸이하'에 비해서는 일에 대한 선호도가 낮다. 중고등학생 때 하던 '수능, 입시를 위한 공부'에 대한 부담이나 스트레스는 조금 경감되어서 일에 대한 선호도가 조금 높아지긴 했지만, 대졸의 경우 대학교 4년 이상까지 공부를 지속해온 상황이라 역시나 일에 대한 선호도가 '대학생, 대학원생'을 넘어서지는 못하는 것이 아닐까 생각이 든다. 어쨋든, 대학까지 다녔다는 것은 취업 외에도 학업을 하며 취업 전에 자기개발을 하고 싶었던 학생들이 많을테니 그 시간을 더 즐기고 취업준비하기 싫은.. 그런 마음... 아닐까...

# 여러 가치관 별로 어떤 상관관계가 있을까

## 이건 2018년 데이터로 분석함.
'학력별'로 각 카테고리에서 어떤 응답 추이를 보이는지 비교해보다보니까 비슷한 개형을 보이는 그래프들도 있고, 좀 다른 개형을 보이는 그래프도 있었음. 비슷한 개형을 보이는 경우에는 상대적으로 해당 가치관들 간의 상관관계가 높은 것이 아닐까 하는 생각이 들었고, 가치관별로 상관관계를 알아볼 수 있는 방법을 고민하였음.  

고민한 결과, 원래 데이터 상에서는 상관관계를 따지기에는 각 가치관 별로 항목이 너무 많이 쪼개져 있어서 분석이 어려웠음. 그래서 나는 설문에서 다루고 있는 '가치'들 별로 대표값 하나로 정의를 하고, 그 값들 사이의 상관관계를 분석해보기로 했음. 즉, '일 vs 여가'에서는 '일을 매우 더 중요시함'-'일을 약간 중요시함'을 '일'이라는 가치에 해당하는 값으로 환산하고, '여가를 약간 중요시함'-'여가를 매우 더 중요시함'을 '여가'라는 가치에 해당하는 값으로 환산하고자 함. '차이없음'은 제외하기로 함.   

이에, 세 가지 응답을 하나의 값으로 어떻게 합칠까 고민하다가 각각에 가중치를 달리 주는 것이 좀더 정확할 것 같아서 '매우 더'를 0.5의 가중치를 주고, '더'를 0.3, '약간'에 0.2의 가중치를 줘서 가중치 총합 1을 맞췄음. 그래서 각각의 응답비율에 가중치를 곱한후에 더한값을 'work'라는 새로운 column에 값을 저장하도록 했음. pandas에서 이미 column간 연산이 자유롭기 때문에 쉽게 코드 구현 가능.   

같은 방식으로 '일', '여가', '현실', '이상', '결과', '과정', '개인', '집단' 각각의 대표값을 산정해낸 후에 correlation graph를 그려봄.

3) 어두운 것중에서 눈에 들어온건 'leisure'와 'reality'간의 상관관계였다. 매우 짙은 색을 보이고 있어 매우 강한 음의 상관관계를 보이고 있음을 확인할 수 있는데, 즉,,, 여가에 대한 선호도, 중요도를 높게 생각하는 사람들이 '현실'에 대한 중요도는 낮게 생각함을 의미한다. 즉,,, '여가'와 '현실'은 거리가 멀다... 공존하기 어렵다... 라는 것... 참고로 '여가'와 '이상'을 보게 되면 아주 밝은 양의 상관관계를 보임을 볼 수 있다. '이상'과 '여가'는 어느정도 비례하지만, '이상'과 '현실'은 반비례한다는 씁쓸한 결과가 인상적이었다.

4) 'leisure'와 'group'도 매우 어두운 색이다. 음의 상관관계를 보인다는 건데,,, 잘 와닿지 않아서 'work'와 'group'이 양의 상관관계 보인다는 것에 초점 맞춰서 생각해보니까 조금 더 와닿았다. '일'을 중요시하는 사람들이 '집단'을 더 중요시하고, '여가'를 중요시하는 사람들은 상대적으로 '개인'을 더 중시한다는 것. 뭔가 나는 이 둘간의 상관관계를 떠올리면 오히려 일할 때가 마이웨이고, 놀때가 집단이 잘 어울린다고 생각했는데 조금은 의외의 결과였다.

```python

df2018 = data2018.loc['중·고등학생':'대졸이상',:]
df2018 = df2018.astype('float')
df2018
```

```python
df2018 = data2018.loc['중·고등학생':'대졸이상',:]
df2018 = df2018.astype('float')
df2018['work'] = df2018['일을 매우 더 중시함']*0.5+df2018['일을 더 중시함']*0.3+df2018['일을 약간 더 중시함']*0.2
df2018['leisure'] = df2018['여가를 매우 더 중시함']*0.5+df2018['여가를 더 중시함']*0.3+df2018['여가를 약간 더 중시함']*0.2
df2018['reality'] = df2018['현실을 매우 더 중시함']*0.5+df2018['현실을 더 중시함']*0.3+df2018['현실을 약간 더 중시함']*0.2
df2018['ideal'] = df2018['이상을 매우 더 중시함']*0.5+df2018['이상을 더 중시함']*0.3+df2018['이상을 약간 더 중시함']*0.2
df2018['result'] = df2018['결과를 매우 더 중시함']*0.5+df2018['결과를 더 중시함']*0.3+df2018['결과를 약간 더 중시함']*0.2
df2018['process'] = df2018['과정을 매우 더 중시함']*0.5+df2018['과정을 더 중시함']*0.3+df2018['과정을 약간 더 중시함']*0.2
df2018['individual'] = df2018['개인을 매우 더 중시함']*0.5+df2018['개인을 더 중시함']*0.3+df2018['개인을 약간 더 중시함']*0.2
df2018['group'] = df2018['집단을 매우 더 중시함']*0.5+df2018['집단을 더 중시함']*0.3+df2018['집단을 약간 더 중시함']*0.2
new_df2018 = df2018.loc[:, 'work':'group']
sns.heatmap(new_df2018.corr())
plt.rcParams['axes.unicode_minus'] = False
plt.title("가치별 상관관계")
plt.xlabel("values")
plt.ylabel("values")
```
