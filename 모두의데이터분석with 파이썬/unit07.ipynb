```python
region = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해주세요:  ")

import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
result = []

for row in data:
    if region in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))

plt.title(region+'지역의 연령대별 인구 분포도')
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.plot(result,color='blue', ls=':')
plt.show()
```
