```python
import csv
f = open('subwaytime202103.csv',encoding='utf8')
data = csv.reader(f)
next(data)
next(data)
station = ['']*24
max_list = [0]*24

for row in data:
    row[4:] = map(int, row[4:])
    for t in range(4):
        num = row[2*t+44]
        if num > max_list[t]:
            max_list[t] = num
            station[t] = row[1]+' '+row[3]+'('+str(t)+')'  # 역 이름 뒤에 시각도 같이 표시하자!
    for t in range(4,24):
        num = row[2*t-4]
        if num > max_list[t]:
            max_list[t] = num
            station[t] = row[1]+' '+row[3]+'('+str(t)+')'

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.title('시간대별 최다 승차 지하철역과 승차 인원')
plt.bar(range(24),max_list,color='hotpink')
plt.xticks(range(24),station,rotation=90) # x축에 시간대별 1위 차지한 역 이름 90도 회전해서 표현하기!
plt.show()
```

## 그냥 내가 궁금해서 해보는 강남역의 시간별 승차인원 수 변화 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime202103.csv',encoding='utf8')
data = csv.reader(f)
next(data)
next(data)
station = input('어떤 지하철역의 시간대별 승차 인원이 궁금한가요?: ')
on_num = [0]*24
off_num = [0]*24

for row in data:
    row[4:]=map(int,row[4:])
    if row[3]!=str(station):
        continue
    else:
        for i in range(4):
            on_num[i] += row[2*i+44]
            off_num[i] += row[2*i+45]
        for i in range(4,24):
            on_num[i] += row[2*i-4]
            off_num[i] += row[2*i-3]

plt.rc('font',family='Malgun Gothic')
plt.style.use('ggplot')
plt.title('시간대별 '+str(station)+'역 승하차인원')
plt.plot(range(24),on_num,label='승차인원', color='red')
plt.plot(range(24),off_num, label='하차인원', color='purple')
plt.legend()
plt.show()            
```

## 아래는 교재에서 실습 문제로 나온 시간대별 승하차인원을 모든 역 합쳐서 정리한 그래프

```python
import csv
import matplotlib.pyplot as plt
f =open('subwaytime202103.csv',encoding='utf8')
data = csv.reader(f)
next(data)
next(data)
on_num = [0]*24
off_num = [0]*24

for row in data:
    row[4:] = map(int,row[4:])
    for i in range(4):
        on_num[i] += row[2*i+44]
        off_num[i] += row[2*i+45]
    for i in range(4,24):
        on_num[i] += row[2*i-4]
        off_num[i] += row[2*i-3]
        
plt.rc('font',family='Malgun Gothic')
plt.style.use('ggplot')
plt.title('시간대별 지하철 승하차인원')
plt.plot(range(24),on_num,label='승차인원', color='purple')
plt.plot(range(24),off_num, label='하차인원', color='green')
plt.legend()
plt.show()
```
