```python
import csv
f = open('subwayfee202103.csv', encoding = 'utf8')
data = csv.reader(f)
next(data)
max_rate = 0
max_line = ''
max_stop_name = ''

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
    rate = row[4]/(row[4]+row[6])
    if rate > max_rate:
        max_rate = rate
        max_line = row[1]
        max_stop_name = row[3]

print('2021년 3월의 유임승차인원 비율은 ',max_line,'의',max_stop_name,'이',max_rate,'로 가장 높았습니다.')
```

```python
import csv
f = open('subwayfee202103.csv', encoding = 'utf8')
data = csv.reader(f)
next(data)
max_rate = 0
max_line = ''
max_stop_name = ''

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
    if row[6] != 0:
        rate = row[4]/(row[4]+row[6])
        if rate>0.93:
            print(row,round(rate,3))
        if rate > max_rate:
            max_rate = rate
            max_line = row[1]
            max_stop_name = row[3]

print('2021년 3월의 유임승차인원 비율은 ',max_line,'의',max_stop_name,'이(가)',round(max_rate,2),'로 가장 높았습니다.')
```

```python
import csv
f = open('subwayfee202103.csv', encoding = 'utf8')
data = csv.reader(f)
next(data)
max_on = 0
max_off = 0
max_on_name = ''
max_off_name = ''

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
        
    on = row[4]+row[6]
    if on>1500000: print('on',row[1],' ',row[3])
    off = row[5]+row[7]
    if off>1500000: print('off',row[1],' ',row[3])
        
    if on > max_on:
        max_on = on
        max_on_name = row[1]+row[3]
    if off > max_off:
        max_off = off
        max_off_name = row[1]+row[3]

print('2021년 3월 승차 인원이 가장 많은 역은',max_on,'명이 승차한',max_on_name,'이고, 하차 인원이 가장 많은 역은',max_off,'명이 하차한',max_off_name,'입니다.')
```
