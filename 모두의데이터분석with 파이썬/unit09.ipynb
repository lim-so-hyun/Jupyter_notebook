```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

male = []
female = []

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:]:
            female.append(int(i))
        break   # 이거 없음 또 제주특별자치도 데이터 여러 줄이라 오류 생기니까 반드시 넣어줘야 함!
        
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name+' 지역의 남녀 성별 인구 분포')
plt.barh(range(101),male,label='남성')
plt.barh(range(101),female, label='여성')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
size=[10,330,54,91]
label=['봄', '여름', '가을', '겨울']
color=['palegreen','lightblue','darksalmon','mediumpurple'] # 데이터 순서 맞춰서 색상 정해주기
plt.axis('equal')  # 파이그래프 완전한 원 형태(타원 아니고)로 그려지게 함
plt.pie(size,labels=label,autopct='%.1f%%', colors=color, explode=(0.1,0,0.2,0)) # lables,autopct,colors 속성 이용해서 지정해주고, explode 속성 이용해서 돌출값 지정해주기
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
size=[]
name = input("찾고 싶은 지역의 이름을 알려주세요: ")

for row in data:
    if name in row[0]:
        male_size = 0
        female_size = 0
        for i in range(101):
            male_size += int(row[i+3])
            female_size += int(row[i+106])
        break
    
size.append(male_size)
size.append(female_size)

plt.rc('font',family='Malgun Gothic')
color=['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남','여'], autopct='%.1f%%', colors=color, startangle=90)  # 원래 3시방향에서 시작해서 반시계방향으로 표현되는데 90도에서 시작하라고 지정!
plt.legend()
plt.title(name+' 지역의 남녀 성별 비율')
plt.show()
```
