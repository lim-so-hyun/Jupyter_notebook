```python
import matplotlib.pyplot as plt
import random
import numpy as np

result = []
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
result = np.array(result)
print('1/4: '+str(np.percentile(result,25)))
print('2/4에 해당하는 값은: '+str(np.percentile(result,50)))
print('3/4: '+str(np.percentile(result,75)))
```

```python
# 1월과 8월의 최저기온 boxplot 그리기
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '08':
            aug.append(float(row[4]))
        if month == '01':
            jan.append(float(row[4]))
            
import matplotlib.pyplot as plt
plt.title('january and august lowest temperature', color='blue')
plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
    
```

```python
# 1월과 8월의 최저기온 boxplot 그리기
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '08':
            aug.append(float(row[4]))
        if month == '01':
            jan.append(float(row[4]))
            
import matplotlib.pyplot as plt
plt.title('january and august lowest temperature', color='blue')
plt.boxplot([jan,aug])  # data를 이런식으로 각각을 원소로 하는 리스트로 넣어주면?
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
monthly_list = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    month = int(row[0].split('-')[1])
    if row[2] != '':
        monthly_list[month-1].append(float(row[2])) # 월수에서 1 빼줘야 index가 됨.
            
import matplotlib.pyplot as plt
plt.title('monthly highest temperature', color='red')
plt.boxplot(monthly_list)  
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)

date_list = []
for _ in range(31):
    date_list.append([])

for row in data:
    month = int(row[0].split('-')[1])
    date = int(row[0].split('-')[2])
    if row[2] != '' and month == 5 :
        date_list[date-1].append(float(row[2])) 
            
import matplotlib.pyplot as plt
plt.title('Highest temperature in May', color='green')
plt.boxplot(date_list)  
plt.show()
```

```python
# 그래프 모양에 옵션 몇 가지 추가!

import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)

date_list = []
for _ in range(31):
    date_list.append([])

for row in data:
    month = int(row[0].split('-')[1])
    date = int(row[0].split('-')[2])
    if row[2] != '' and month == 5 :
        date_list[date-1].append(float(row[2])) 
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')  # 배경이 회색의 격자무늬, 2/4값 나타내는 선이 파란색인 style로 지정
plt.figure(figsize=(20,5), dpi=300) # 그래프 크기 수정
plt.title('Highest temperature in May', color='green')
plt.boxplot(date_list, showfliers=False)  # 아웃라이어 값들을 보이지 않게 하겠다는 option 추가

plt.show()
```
