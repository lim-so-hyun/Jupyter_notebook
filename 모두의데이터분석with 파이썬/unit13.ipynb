# numpy의 sub라이브러리_ random 서브 라이브러리_`rand()` 함수
`rand()` 함수 사용하면 0~1 사이에 있는 n개의 실수가 랜덤하게 생성됨  
`randint()` 함수의 실수 버전 느낌!  

`choice()` 함수는 특정 range에서 몇개의 숫자를 뽑으라는 것. replace=False 추가하면 한번 뽑은 숫자 다시 뽑지 못하게 할 수 있음.

```python
import numpy as np
a = np.random.rand(5)
print(a)
print(type(a))
```

```python
import numpy as np
a = np.random.choice(7,10)  # range(7)에 해당하는 값들로 10개를 추출해라
print(a)
print(type(a))
```

```python
import numpy as np
a = np.random.choice(10,6,replace=False)
print(a)
```

```python
import numpy as np
a = np.random.choice(6,10,p=[0.1,0.2,0.3,0.2,0.1,0.1])  # 지정한 범위의 각 값이 나올 확률 지정. 적은 숫자로 했을 때에는 정확히 맞지 않는데 뽑는 수 늘리면 근접해짐
print(a)
```

# numpy 이용해서 그래프 그리기

# numpy 이용한 버블 차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,100,200)  # 10 이상 100 미만의 정수 200개 추출
y = np.random.randint(10,100,200)
size = np.random.rand(200) * 100  # 0과 1사이의 실수 100개를 추출하고 이에 100을 곱해준 것

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

## 아래 예제 mask 적용하면 계속 오류 뜨는 데 어떻게 수정해야할지 모르겠다 ㅠㅠ

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)
size = np.random.rand(1000) *100
mask1 = abs(x) > 50
mask2 = abs(y) > 50

plt.scatter(x, y, s=size, c=x, cmap='jet',alpha=0.7) # x와 y가 각각 random한 숫자들로 구성되어있다보니, mask를 적용하면 남은 갯수가 달라지면서 x,y 갯수가 달라지고 size 갯수도 안맞아져서 오류 뜨는 듯
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-100,100,1)
y = np.arange(-200,200,2)
size = np.random.rand(200) *100
mask1 = abs(x) % 2 == 0
mask2 = abs(y) % 2 == 0

plt.scatter(x, y, s=size, c=x, cmap='jet',alpha=0.7) 
plt.colorbar()
plt.show()
```

교재에서 주어진 error가 계속 x, y, size 갯수 안맞아서 뜨는 건가 해서 random 대신 내가 아는 배열 가지고 mask 적용해서 그려봤음.  
그런데 이렇게 하는 거면,,, random에서는 mask 설정하는 순간 몇개가 그 mask 에 해당하는지 모르니까 size 배열을 설정할 수가 없을 텐데. size 뿐 아니라 x, y 갯수 일치시키는 것도 불가능.  

제대로 된 mask 기능이면 mask 적용안한 상태에서 그래프 그린다고 생각하고 그리고 mask에 해당하는 부분만 남기고 나머지를 가리는 식으로 이루어져야할 것 같은데 자꾸 에러 뜨는 걸로 봐서는 mask를 이미 적용한 후에 x, y를 다시 update한 후에 그래프를 그리려고 하니까 애초에 그래프가 그려지는 조건 만족시키는 게 너무 번거로움.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-100,100,1)
y = np.arange(-200,200,2)
size = np.random.rand(99) *100
mask1 = abs(x) > 50
mask2 = abs(y) > 100
x = x[mask1]
y = y[mask2]
plt.style.use('ggplot')
plt.scatter(x, y, s=size, c=x, cmap='jet',alpha=0.7) 
plt.colorbar()
plt.show()
```
