```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
result = []
region = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해주세요:  ")  # 이걸 맨 앞으로 뺴면 region변수가 아래까지 반영이 안되어서 이상하게 됨. 

for row in data:
    if region in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))

plt.title(region+' 지역의 연령대별 인구 분포도')
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.plot(result,color='blue', ls=':')
plt.show()  # 제대로 했는데, 동 입력 시에 같은 동이 여러 개 있으면 ex. 수지구 그래프 이상하게 나옴. 성복동처럼 하나만 있을 때는 문제 없음.
```

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
result = []
region = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해주세요:  ")  

for row in data:
    if region in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))
        break   # 여기에 break문 추가하면 같은 지역 이름이 여러 열에 있더라도, 첫번째로 포함되어있는 열에 대해서 데이터 출력하고 break되므로 그래프 정상적으로 나옴.
                 # 해당 지역 첫번쨰로 나오는 열이 그 지역 총합한거 정리된 데이터라서 그거 출력하면 우리가 알고자 하는 해당 인구 분포 정확히 아는 것 맞음! 굳이 그 지역 이름 포함된 데이터 다 뽑아서 더할 필요 없음.

plt.title(region+' 지역의 연령대별 인구 분포도')
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.plot(result,color='blue', ls=':')
plt.show()  
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

male = []
female = []
for row in data:
    if '성복' in row[0]:
        for i in range(3,104):
            male.append(-int(row[i]))
        for i in range(106,207):
            female.append(int(row[i]))  

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False   # 마이너스 기호 안깨지게 하기 위한 코드
plt.title('성복역 지역의 남녀 성별 인구 분포')
plt.barh(range(101),male, label='남성', color='yellow')  # 남성이라는 범례 추가
plt.barh(range(101),female, label='여성', color='pink')  # 여성이라는 범례 추가
plt.legend()  # 범례 보여줘라
plt.show()            
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

male = []
female = []
region = input('성별에 따른 연령대별 인구수를 알고자 하는 지역의 이름을 입력해주세요: ')

for row in data:
    if region in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:]:
            female.append(int(i))  
        break  #여기서도 이거 중요! region 이른 두 번째 나오기 전에 break 시켜줘야 함

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False   # 마이너스 기호 안깨지게 하기 위한 코드
plt.title(region+' 지역의 남녀 성별 인구 분포')
plt.barh(range(101),male, label='남성',color='green')  
plt.barh(range(101),female, label='여성', color='hotpink')  
plt.legend()  # 범례 보여줘라
plt.show()            
```
