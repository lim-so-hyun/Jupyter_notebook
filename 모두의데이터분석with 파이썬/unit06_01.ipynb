```python
import matplotlib.pyplot as plt
import random
dice=[]
for i in range(10):
    dice.append(random.randint(1,6))

plt.hist(dice, bins=6)  # dice는 사용할 자료, bins는 가로축 구간 개수 설정
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
dice=[]
for i in range(100000):  # 시행횟수 큰 수로 늘리면 각 숫자 나올 확률 1/6으로 수렴
    dice.append(random.randint(1,6))

plt.hist(dice, bins=6)  # dice는 사용할 자료, bins는 가로축 구간 개수 설정
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[2] != '':
        result.append(float(row[2]))

plt.hist(result,bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data = csv.reader(f)
next(data)
august = []

for row in data:
    if row[2]!='':
        if int(row[0].split('-')[1])==8:
            august.append(float(row[2]))
plt.title('August highest temperature')
plt.hist(august,bins=100, color='r')
plt.show()
        
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data = csv.reader(f)
next(data)
august = []
feburary = []

for row in data:
    if row[2]!='':
        if int(row[0].split('-')[1])==8:
            august.append(float(row[2]))
        if int(row[0].split('-')[1])==1:
            feburary.append(float(row[2]))
plt.title('Feb and Aug highest temperature')
plt.hist(august,bins=100, color='red')
plt.hist(feburary,bins=100, color='blue')
plt.show()
```
