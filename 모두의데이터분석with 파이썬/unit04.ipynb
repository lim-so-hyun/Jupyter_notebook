```python
import matplotlib.pyplot as plt
plt.title('color practice')
plt.plot([10,20,30,40], color='g', label = 'green')
plt.plot([40,30,10,2],color='r', label = 'red')
plt.legend(loc=7)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('line style modification practice')
plt.plot([1,2,3,4,5], color='r', linestyle='--', label='dashed')
plt.plot([2,3,4,5,6], 'b::', label = 'dotted')
plt.plot([5,4,3,2,1], 'y', label='original')
plt.legend(loc=7)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('line style modification practice')
plt.plot([1,2,3,4,5], color='r', linestyle='--', label='dashed')
plt.plot([2,3,4,5,6], 'b',ls=':', label = 'dotted')
plt.plot([5,4,3,2,1], color='y', label='spotted')
plt.plot([6,5,4,3,2], 'k--', label='black dashed')
plt.legend(loc=7)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker shape modification practice')
plt.plot([1,2,3,4,5], color='r', label='original')
plt.plot([2,3,5,7,8], 'k^', ls='--', label='dashed triangle')
plt.plot([10,6,5,4,3], 'g.', label='green dot')
plt.plot([11,8,7,5,4], 'y.:', label='yellow dot line')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker shape modification practice')
plt.plot([1,2,3,4,5], color='r', label='original')
plt.plot([2,3,5,7,8], 'k^', ls='--', label='dashed triangle')
plt.plot([10,6,5,4,3], 'g.', label='green dot')
plt.plot([11,8,7,5,4], 'y.:', label='yellow dot line')
plt.plot([10,10,10,10,10],'^',label='random color triangle')
plt.legend()
plt.show()
```
