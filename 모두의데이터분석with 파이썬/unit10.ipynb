# 지역의 연령대별 남녀 인구 비율 시각화하기

## 꺾은선 그래프로 표현하기

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
```

```python
name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break
```

```python
import matplotlib.pyplot as plt
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

## 막대그래프로 표현하기
연령별로, 남성 인구수에서 여성 인구수를 뺀 차이를 막대그래프로 표현 가능

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
male_minus_female = []

name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            male_minus_female.append(int(row[i])-int(row[i+103]))
        break
        
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False # 마이너스 기호 안깨지게 하기
plt.title(name+'지역의 연령대별 남녀 인구 차이')
plt.barh(range(101), male_minus_female, color='violet')
plt.show()
```

# 산점도
``scatter()`` 함수 사용

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')   # 뒤에 격자무늬 있는 스타일 사용하고 싶어서
plt.scatter([1,2,3,4],[10,6,20,34]) # 앞에 넣어준게 x축, 뒤에 넣은게 y축
plt.show()
```

# 버블 차트
산점도와 동일하게 ``scatter()`` 함수를 사용하는데, 기존 코드에서 **size**를 의미하는 **s**속성을 추가해주고, 원하는 크기 입력 필요

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')   # 뒤에 격자무늬 있는 스타일 사용하고 싶어서
plt.scatter([1,2,3,4],[10,6,20,34],s=[5,10,50,20]) # s 속성으로 점 크기 지정
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')   # 뒤에 격자무늬 있는 스타일 사용하고 싶어서
plt.scatter([1,2,3,4],[10,6,20,34],s=[5,10,50,20],c=['red','blue','green','gold']) # c 속성 추가해서 색도 지정
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.scatter([1,2,3,4],[10,30,40,80], s=[200,50,500,800],c=range(4),cmap='jet') # cmap 속성 추가해서 색상 종류 설정 가능함. jet은 무지개색 jet 컬러맵
plt.colorbar()  # colorbar 보이기 위한 것. 이거 없어도 버블 차트 나오기는 하는데 어떤 색이 어떤 값에 해당하는지 모르니까 보이게 하자
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []
for i in range(100):  # 이번에는 100개 무작위 데이터로 그려보자
    x.append(random.randint(50,100)) # 50-100 사이 무작위 숫자
    y.append(random.randint(50,100))
    size.append(random.randint(10,100)) # 크기도 random
plt.scatter(x,y,s=size)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []
for i in range(100):  
    x.append(random.randint(50,100)) 
    y.append(random.randint(50,100))
    size.append(random.randint(10,100)) 
plt.scatter(x,y,s=size,c=size,cmap='gnuplot2',alpha=0.5)
# size에 따라 color도 지정되도록 c 속성에다가 size 넣어줌.
# alpha 속성 조정하면 투명도 설정 가능. 작은 원도 가려지지 않게
plt.colorbar()
plt.show()
```

# 특정 지역의 연령대별 성별 비율 산점도로 표현하기

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

name = input('어느 동네 인구 분포가 알고 싶으신가요?: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break

import matplotlib.pyplot as plt
plt.scatter(m,f,c=range(101),alpha=0.5,cmap='gnuplot2')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')  # 꺾은선 그래프 함수 이용해서 추세선 추가하기
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = [] # 남녀 수 합친 size 자료 만들어서 이걸로 점 크기 설정하자.
        # 근데 그래프 출력해보면 알겠지만 이렇게 해서 s에 넣으면 원 크기 너무너무 큼ㅋㅋㅋㅋㅋㅋㅋㅋ

name = input('어느 동네 인구 분포가 알고 싶으신가요?: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(int(row[i])+int(row[i+103]))
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5),dpi=300) # 크기 설정
plt.title(name+'지역의 성별 인구 그래프')
plt.scatter(m,f,s=size, c=range(101),alpha=0.5,cmap='gnuplot2')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')
plt.xlabel('남성 인구수')  # x축, y축 label 쉽게 추가하기
plt.ylabel('여성 인구수')
plt.show()
```

```python
import csv
import math  # 뒤에서 제곱근 함수 쓰기 위해서 import

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = [] # 남녀 수 합친 size 자료 만들어서 이걸로 점 크기 설정하자.

name = input('어느 동네 인구 분포가 알고 싶으신가요?: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i])+int(row[i+103]))) # 제곱근 씌워서 크기 좀 낮추기
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5),dpi=300) # 크기 설정
plt.title(name+'지역의 성별 인구 그래프')
plt.scatter(m,f,s=size, c=range(101),alpha=0.5,cmap='gnuplot2')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')
plt.xlabel('남성 인구수')  # x축, y축 label 쉽게 추가하기
plt.ylabel('여성 인구수')
plt.show()
```
