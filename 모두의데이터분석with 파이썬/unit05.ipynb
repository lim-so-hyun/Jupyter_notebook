```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[2] != '':
        if row[0].split('-')[1]=='05' and row[0].split('-')[-1]=='11':   #매년 5월 11일 데이터만 가지고 그려보자!
            result.append(float(row[2]))
plt.figure(figsize = (30,2))
plt.plot(result,'blue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []   # 최고기온 담을 list
low = []   # 이번엔 최저기온도 그려보려고 list 따로 만듦.

for row in data:
    if row[2] != '' and row[4] != '':  # 이번에는 최저, 최고 기온 모두 데이터가 제대로 되어있는지 확인 필요
        if 1999 <= int(row[0].split('-')[0]):  
            if row[0].split('-')[1] == '05' and row[0].split('-')[-1]=='11':
                high.append(float(row[2]))
                low.append(float(row[4]))
plt.figure(figsize = (30,2))
plt.plot(high,'red')
plt.plot(low,'blue')
plt.show()                
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []   # 최고기온 담을 list
low = []   # 이번엔 최저기온도 그려보려고 list 따로 만듦.

for row in data:
    if row[2] != '' and row[4] != '':  # 이번에는 최저, 최고 기온 모두 데이터가 제대로 되어있는지 확인 필요
        if 1999 <= int(row[0].split('-')[0]):  
            if row[0].split('-')[1] == '05' and row[0].split('-')[-1]=='11':
                high.append(float(row[2]))
                low.append(float(row[4]))
plt.figure(figsize = (30,2))
plt.title('0511 highest and lowest temp since 1999')
plt.plot(high,'red', ls=':',label = 'highest temp')
plt.plot(low,'blue', label = 'lowest temp')
plt.legend(loc=7)  # 범례 표시
plt.show()  
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []   # 최고기온 담을 list
low = []   # 이번엔 최저기온도 그려보려고 list 따로 만듦.

for row in data:
    if row[2] != '' and row[4] != '':  # 이번에는 최저, 최고 기온 모두 데이터가 제대로 되어있는지 확인 필요
        if 1999 <= int(row[0].split('-')[0]):  
            if row[0].split('-')[1] == '05' and row[0].split('-')[-1]=='11':
                high.append(float(row[2]))
                low.append(float(row[4]))
plt.figure(figsize = (30,2))
plt.rc('font',family='Malgun Gothic')  # 한글은 글자 깨질 수 있어서 글씨체 기본 글꼴 설정
plt.rcParams['axes.unicode_minus']=False # 만약 세로축 값에 마이너스 등장하면 마이너스 기호 깨질 수 있어서 기호 깨짐 방지하는 문장
plt.title('1999년 이후 내 생일날의 기온 변화 그래프')  # 이번엔 한글 제목 설정
plt.plot(high,'hotpink', ls=':',label = 'highest temp')
plt.plot(low,'skyblue', label = 'lowest temp')
plt.legend(loc=7)  # 범례 표시
plt.show()  
```
